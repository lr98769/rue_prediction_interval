# Preprocessing S&P500 Data

```python
import os
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import joblib
import pickle 

fp_data = os.path.join("../../", "data")
fp_preprocessed = os.path.join("../../", "processed_data")
fp_data_df = os.path.join(fp_data, "all.csv")
fp_recession_df = os.path.join(fp_data, "bus_dates.csv")
fp_preprocessed_data_dict = os.path.join(fp_preprocessed, "snp500_data_dict.joblib")
fp_scaler = os.path.join(fp_preprocessed, "snp500_scaler.pkl")

feat_cols = ["U$", "Mkt-RF", "SMB", "HML", "Mom"]
pred_col = "U$"
date_col = "Date"
peak_col, trough_col = "Peak month", "Trough month"
```

## Load Data

```python
df = pd.read_csv(fp_data_df)
df[date_col] =  pd.to_datetime(df[date_col], format="%Y-%m-%d")
df
```

```python
recession_df = pd.read_csv(fp_recession_df)
recession_df[peak_col] =  pd.to_datetime(recession_df[peak_col], format="%Y-%m-%d")
recession_df[trough_col] =  pd.to_datetime(recession_df[trough_col], format="%Y-%m-%d")
recession_df
```

## Add Recession Information

```python
def add_recession_info(df, recession_df):
    df = df.copy()
    n = len(df)
    recession_mask = np.repeat(False, n)
    for i in range(len(recession_df)):
        peak, trough = recession_df.iloc[i]
        cur_mask = np.logical_and((df[date_col] > peak), (df[date_col] < trough))
        recession_mask = np.logical_or(recession_mask, cur_mask)
    df["recession"] = recession_mask

    return df 
df = add_recession_info(df, recession_df)
print(df["recession"].value_counts())
df
```

## Split Records into Train, Valid, Test Records

```python
def split_df(df, valid_prop, test_prop):
    n = len(df)
    n_train, n_valid, n_test = round(n*(1-valid_prop-test_prop)), round(n*valid_prop), round(n*test_prop)
    df["split"] = np.concatenate([np.repeat("Train", n_train), np.repeat("Valid", n_valid), np.repeat("Test", n_test)])
    return df

df = split_df(df, valid_prop=0.1, test_prop=0.2)
display(df)
print(df["split"].value_counts())
```

### Show Split

```python
def show_split(df, pred_col, recession_df):
    split_col = "split"
    date_col = "Date"
    df = df.copy()

    # Plot Target
    fig, ax = plt.subplots(1, 1, figsize=(5, 2), dpi=300)
    df[date_col] = pd.to_datetime(df[date_col])
    ax.plot(df[date_col], df[pred_col])

    # Add recession information
    alpha = 0.3
    for i in range(len(recession_df)):
        start, end = recession_df.iloc[i]
        ax.axvspan(
            xmin=start,
            xmax=end,
            ymin=0, ymax=1,
            alpha=alpha,
            label="Recession" if i == 0 else None,
            color="#adadad",
        )

    # Get split intervals
    split_intervals = {}
    for split_name, split_df in df.groupby(split_col):
        split_intervals[split_name] = [
            split_df.iloc[0][date_col], split_df.iloc[-1][date_col],
        ]
    num_splits = len(split_intervals)

    # Add split information
    color_map = {
        "Train": "#70D6FF",
        "Valid": "#FF9770",
        "Test": "#FFD670"
    }
    splits = ["Train", "Valid" , "Test"]
    for i, split in enumerate(splits):
        start, end = split_intervals[split]
        ax.axvspan(
            xmin=start,
            xmax=end,
            ymin=1/num_splits*i,
            ymax=1/num_splits*(i+1),
            alpha=alpha,
            label=split,
            color=color_map[split],
        )
    
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=4)
    plt.tight_layout()
    

show_split(df, pred_col, recession_df)
plt.savefig("s&p500_data_split.jpg", bbox_inches="tight")
```

## Standardise Data

```python
def standardise_data(df, feat_cols):
    df = df.copy()
    train_df = df[df["split"]=="Train"]
    valid_df = df[df["split"]=="Valid"]
    test_df = df[df["split"]=="Test"]
    scaler = StandardScaler()
    scaler.fit(train_df[feat_cols])
    train_df.loc[:, feat_cols] = scaler.transform(train_df.loc[:, feat_cols])
    valid_df.loc[:, feat_cols] = scaler.transform(valid_df.loc[:, feat_cols])
    test_df.loc[:, feat_cols] = scaler.transform(test_df.loc[:, feat_cols])
    return pd.concat([train_df, valid_df, test_df]), scaler

df_scaled, scaler = standardise_data(df, feat_cols)
df_scaled
```

```python
with open(fp_scaler, 'wb') as handle:
    pickle.dump(scaler, handle)
```

## Flatten Data and Generate Data for Training

```python
def flatten(df: pd.DataFrame, feat_cols: list, pred_col: str, feat_days: int, pred_days: int):
    df = df.copy()
    n = len(df)
    date_col = "Date"
    split_col = "split"
    recession_col = "recession"
    df[date_col] = df[date_col].dt.strftime("%Y-%m-%d")
    # Flattened df columns
    flattened_feat_cols = [f"{feat_col}_day{i}" for i in range(feat_days) for feat_col in feat_cols]
    flattened_pred_cols = [f"{pred_col}_t+{i+1}" for i in range(pred_days)]
    flattened_split_cols = [f"{split_col}_t+{i+1}" for i in range(pred_days)]
    flattened_recession_cols = [f"{recession_col}_t+{i+1}" for i in range(pred_days)]
    flattened_date_cols  = ["input_start date", "input end date"]   + [f"target_date_t+{i+1}" for i in range(pred_days)]
    flattened_index_cols = ["input start index", "input end index"] + [f"target_index_t+{i+1}" for i in range(pred_days)]
    all_cols = flattened_feat_cols + flattened_pred_cols + flattened_date_cols + flattened_index_cols + flattened_split_cols + flattened_recession_cols
    # E.g. [0, 1, 2, 3, 4], 5, 6, (7) if feat_days = 5, pred_days = 3
    first_end_index = feat_days + pred_days - 1 
    df_rows = []
    # Iterate through possible end 
    possible_ends = list(range(first_end_index, n))
    for end in tqdm(possible_ends):
        feat_indices = np.arange(end - pred_days - feat_days + 1, end - pred_days + 1)
        pred_indices = np.arange(end - pred_days + 1, end + 1)
        # Create feature set
        feat_vector = df.iloc[feat_indices][feat_cols].values.flatten().tolist()
        # Create prediction set
        pred_vector = df.iloc[pred_indices][pred_col].values.tolist()
        # Input start date
        input_start_date = df.iloc[feat_indices[0]][date_col]
        # Input end date
        input_end_date = df.iloc[feat_indices[-1]][date_col]
        # Target_dates
        target_dates = df.iloc[pred_indices][date_col].values.tolist()  
        date_vector = [input_start_date, input_end_date] + target_dates
        # Input, end and target indices
        index_vector = [feat_indices[0], feat_indices[-1]] + pred_indices.tolist()
        # Split label
        split_vector = df.iloc[pred_indices][split_col].values.tolist()
        # Recession label
        recession_vector = df.iloc[pred_indices][recession_col].values.tolist()
        # Form row vector
        row_vector = feat_vector + pred_vector + date_vector + index_vector + split_vector + recession_vector
        df_rows.append(row_vector)
    df = pd.DataFrame(df_rows, columns=all_cols)
    return df
        
df_flattened = flatten(df = df_scaled, feat_cols = feat_cols, pred_col = pred_col, feat_days = 5, pred_days = 3)
df_flattened
```

## Split Df

```python
def split_df(df, feat_cols, pred_col, feat_days = 5, pred_days = 3):
    df = df.copy()
    split_col_label = "split"
    target_date_label = "target_date"
    target_index_label = "target_index"
    recession_label = "recession"
    flattened_feat_cols = [f"{feat_col}_day{i}" for i in range(feat_days) for feat_col in feat_cols]
    flattened_date_cols  = ["input_start date", "input end date"]
    flattened_index_cols = ["input start index", "input end index"]
    data_dict = {}
    for time_i in range(pred_days):
        cur_dict = {}
        time = time_i + 1
        time_label = f"t+{time}"
        split_col = split_col_label+"_"+time_label
        cur_pred_col = pred_col+"_"+time_label
        cur_target_date_col = target_date_label+"_"+time_label
        cur_target_index_col = target_index_label+"_"+time_label
        cur_recession_col = recession_label+"_"+time_label
        for split_name, split_df in df.groupby(split_col):
            cols = flattened_date_cols+[cur_target_date_col]+flattened_index_cols+[cur_target_index_col]+flattened_feat_cols+[cur_pred_col]+[cur_recession_col]
            cur_dict[f"{split_name.lower()}_df"] = split_df[cols].copy()
        cur_dict["outputs"] = [cur_pred_col]
        data_dict[time_label] = cur_dict
    return data_dict

def display_data_dict(data_dict):
    len_df, time_labels = [], []
    splits = ["train_df", "valid_df", "test_df", "outputs"]
    for time_label, time_info_dict in data_dict.items():
        print(time_label)
        stat_dict = {}
        for split_df_name in splits:
            print(f"- {split_df_name}:")
            split_df = time_info_dict[split_df_name]
            if split_df_name == "outputs":
                stat_dict["pred_col"] = split_df[0]
                continue
            display(split_df)
            stat_dict[split_df_name] = len(split_df)
        len_df.append(stat_dict)
        time_labels.append(time_label)
    len_df = pd.DataFrame(len_df, index=time_labels)
    display(len_df)

data_dict = split_df(df = df_flattened, feat_cols = feat_cols, pred_col = pred_col, feat_days = 5, pred_days = 3)
display_data_dict(data_dict)
```

```python
display_data_dict(data_dict)
```
